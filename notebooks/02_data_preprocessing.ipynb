# Data Preprocessing

This notebook prepares the Geldium credit dataset for predictive modeling by handling missing values, validating data quality, detecting outliers, and exporting a clean dataset suitable for feature engineering and machine learning.

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_excel("data/raw/Delinquency_prediction_dataset.xlsx")

df.head()
```

```python
print("="*60)
print("Dataset Shape")
print("="*60)

print(df.shape)

print()

print("="*60)
print("Data Types")
print("="*60)

df.dtypes
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

missing = missing[missing["Missing Values"] > 0]

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
median_columns = [
    "Income",
    "Loan_Balance",
    "Credit_Score"
]

imputer = SimpleImputer(strategy="median")

df[median_columns] = imputer.fit_transform(df[median_columns])
```

```python
df[median_columns].isnull().sum()
```

```python
df.to_csv(
    "../data/processed/cleaned_data.csv",
    index=False
)
```
